```python
class Car:
    def __init__(self, brand, model): 
        self.brand = brand 
        self.model = model 

    def display(self): 
        return f"Car: {self.brand} {self.model}"


# Ussage 
c = Car("Honda", "Activa")

print(c.display())
```

```python
class Animal: 
    def sound(self): 
        return "Some generic sound"

class Dog(Animal): 
    def sound(self): 
        return "Woof Woof"


# usage 
a = Animal()
d = Dog()

print(a.sound())
print(d.sound())
```
